```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def img_show(img):
    fig = plt.figure()
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    fig.savefig('test', bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
img = cv2.imread('datasets/4A876.png')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_show(img)
```

```python
img = cv2.imread('datasets/4A876.png', 0)
kernel = np.ones((1,1), np.uint8)
```

```python
erosion = cv2.erode(img, kernel, iterations=1)
img_show(erosion)
```

```python
img_list = list()
for i in erosion:
    for j in i:
        img_list.append(j)
for i, j in enumerate(img_list):
    if j < 120:
        img_list[i] = 0
    else:
        img_list[i] = 255
img_new = np.asarray(img_list)
img_new = img_new.reshape(120, 400)
img_show(img_new)
```
