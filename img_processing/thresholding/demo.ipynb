# Thresholding

```python
%matplotlib inline
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
def img_show(img):
    plt.imshow(img, 'gray')
    plt.xticks([])
    plt.yticks([])
    plt.show()
```

```python
def img_save(img, img_name):
    fig = plt.figure()
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    file_name = 'datasets/' + str(img_name)
    fig.savefig(file_name, bbox_inches='tight', pad_inches=0)
    plt.show()
```

```python
img = cv.imread('datasets/4A876.png', 0)
img_show(img)
```

```python
ret, thresh = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
img_show(thresh)
```

```python
img_save(thresh, 'result3')
```

```python
ret, thresh = cv.threshold(img, 127, 255, cv.THRESH_BINARY_INV)
img_show(thresh)
```

```python
ret, thresh = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
img_show(thresh)
```

```python
ret, thresh = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)
img_show(thresh)
```

```python
ret, thresh = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV)
img_show(thresh)
```

```python
img = cv.medianBlur(img, 5)
img_show(img)
```

```python
ret, thresh = cv.threshold(img, 120, 255, cv.THRESH_BINARY)
img_show(thresh)
```

```python
thresh = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
img_show(thresh)
```

```python
thresh = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
img_show(thresh)
```
